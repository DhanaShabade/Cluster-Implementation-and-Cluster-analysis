```python
x = df.iloc[:, [1,2]]
```

```python
km = KMeans(n_clusters=4)
```

```python
ym = km.fit_predict(x)
```

```python
centroids = km.cluster_centers_
centroids
```

```python
plt.figure(figsize=(12, 9))
plt.scatter(x.iloc[:,0], x.iloc[:,1], c = km.labels_)

for i in range(len(np.unique(km.labels_))):
    
    plt.scatter(centroids[i][0], centroids[i][1],marker = "*",c = "r", s = 400)
```

```python
dbscan = DBSCAN(eps=2.5, min_samples=12)

dbscan.fit(x)

labels = dbscan.labels_
np.unique(labels)
```

```python
plt.figure(figsize=(20,15))
plt.scatter(x.iloc[:, 0], x.iloc[:, 1], c=labels)
plt.show()
```

```python
agg_clustering = AgglomerativeClustering()

agg_clustering.fit(x)

labels1 = agg_clustering.labels_
np.unique(labels1)
```

```python
plt.figure(figsize=(20, 15))
plt.scatter(x.iloc[:, 0], x.iloc[:, 1], c = labels1)
plt.show()
```

```python
agg_clustering = AgglomerativeClustering(n_clusters=4)

agg_clustering.fit(x)

labels1 = agg_clustering.labels_
np.unique(labels1)
```

```python
plt.figure(figsize=(20, 15))
plt.scatter(x.iloc[:, 0], x.iloc[:, 1], c = labels1)
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage

Z = linkage(x, 'ward')
```

```python
plt.figure(figsize=(20,15))
dendrogram(Z, leaf_font_size=6, leaf_rotation=90, p = 4, truncate_mode="level")
plt.show()
```
